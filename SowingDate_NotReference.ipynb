```python
def CCDC(t, c0, c1, a1, b1, a2, b2, a3, b3):
    return c0 + c1*t \
        + a1*np.cos(2*np.pi*1*t/T) + b1*np.sin(2*np.pi*1*t/T) \
        + a2*np.cos(2*np.pi*2*t/T) + b2*np.sin(2*np.pi*2*t/T) \
        + a3*np.cos(2*np.pi*3*t/T) + b3*np.sin(2*np.pi*3*t/T)
```
